```python
df = pd.read_csv("covid_19_india.csv",parse_dates=["Date"],dayfirst=True)
```

```python
# Selecting the specific columns
df = df[["Date", "State/UnionTerritory", "Cured","Deaths","Confirmed"]]
```

```python
# Rename the columns
df.columns = ["date","state","cured","deaths","confirmed"]
```

```python
# Nowthe column have been renamed
df.head()
```

```python
df.tail()
```

```python
today=df[df.date == "2020-07-17"]
```

```python
today.head()
```

```python
# maximum confirmed cases in india
max_confirm=today.sort_values(by="confirmed",ascending=False)
```

```python
max_confirm.head()
```

```python
top_5_confirm = max_confirm[0:5]
```

```python
top_5_confirm
```

```python
# Making bar plot using seaborn library---
sns.set(rc={"figure.figsize" :(8,8)})
sns.barplot(x="state", y="confirmed",data=top_5_confirm,hue="state")
plt.show()
```
